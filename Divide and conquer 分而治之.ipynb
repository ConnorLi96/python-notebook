```python
import random

c = ["case1 - build a calculator to evaluate your business model",
"case2 - build a automatic earthquake robot to broadcast the new earthquake",
"case3 - evaluate social media performance of a luxury brand"
]

s = [
18421111,
18421112,
18421113,
18421114,
18421115,
18421116,
18421117,
18421118,
18421119,
18421120,
18421121,
18421122,
18421123,
18421124,
18421125,
18421126,
18421127,
18421128,
18421129,
18421130,
]




random.shuffle(s)
random.shuffle(c)
# a是它们的商数也就是每个case里有多少个学生， b是余数 也就是还有多少个学生可以剩下来。

a = len(s) // len(c) # 这个和int是一回事
b = len(s) % len(c)
i = 0
for i in range(len(c)):  # 在这里把固定的数字换成变量 即便学生和case数量变化也不会出现index error .out of range.
    group = [c[i]]
    stu = s[a*i : a*(i+1)]
    
    if b == 0: 
        print("Assigned case:{}".format(group))
        print("student list:",stu) # 这一步只是解决了当余数等于0的情况
        continue
        # 要让每一个group里面都可以分到一个多出来的学生 
       
    if i < len(c)-(len(c)-b): # 这里取得是 case的数量减去case和余数的差
        stu = s[a*i : a*(i+1)+1] # 要让它打印b遍 还需要找到一个停止的点。
        print("Assigned case:{}".format(group))
        print("student list:",stu)
        print("===========")
  
    else: 
        stu = s[a*i : a*(i+1)]  # 要找运行的次数 当最后到 len(c)-b 组的时候 使用原来的代码
         
        print("Assigned case:{}".format(group))
        print("student list:",stu)
        print("===========")
        
    
            
    i += 1
'''
总结一下：

最后的思路在于，当余数等于0，意味着直接整除打印就行了。但是，当余数不等于0的时候，首先依次打印出：前面的组（比整除情况下，多一个分配学生）。然后再后面的按照整除数量分配的组。
比如:100个学生，13组，余数是9，每组学生整除下来应该是7人。那么先打印前9组（每组8个）。然后打印后4组（每组7个）。


这里着重需要注意的有以下几点
1 变量
要使不论学生和case的数量如何变化，都能够尽量平均分配。在大多数地方都要使用变量。上午的code就有点偷懒的嫌疑。

2 索引
因为case不确定，所以打印的行数也不确定，之前print很多行肯定是不行的。也要使用变量去完成。

3 条件表达式
在遇到这种比较复杂的情况，优先使用条件表达式，真的有助于理清思路。慢慢找到感觉。

另：在找不到思路，或者比较混乱的时候，用#写下来，一点点理清楚。
'''

```
